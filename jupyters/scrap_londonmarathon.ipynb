```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import re

options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

url = "https://results.tcslondonmarathon.com/2024/?page=1&event=MAS&num_results=1000&pid=search&search%5Bsex%5D=%25&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex"
driver.get(url)
time.sleep(4)

resultados = []

# Solo primeros 10 atletas
atletas = driver.find_elements(By.CSS_SELECTOR, "li.list-group-item.row")[:10]

for atleta in atletas:
    try:
        place_overall = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-secondary").text
    except:
        place_overall = "Sin posición general"

    try:
        place_gender = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-primary").text
    except:
        place_gender = "Sin posición género"

    try:
        name_pais = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-fullname a").text
        nombre = re.sub(r"\s*\([A-Z]{3}\)", "", name_pais).strip()
        pais_match = re.search(r"\(([A-Z]{3})\)", name_pais)
        pais = pais_match.group(1) if pais_match else "Sin país"
    except:
        nombre = "Sin nombre"
        pais = "Sin país"

    try:
        categoria = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-age_class").text.replace("Category", "").strip()
    except:
        categoria = "Sin categoría"

    # Evento Mass (asignado directamente)
    event = "Mass"

    # ✅ FINISH con innerHTML + regex (FUNCIONA CON TU HTML)
    try:
        finish_block = atleta.find_element(By.XPATH, ".//div[contains(@class, 'type-time') and .//div[text()='Finish']]")
        html = finish_block.get_attribute("innerHTML")
        match = re.search(r'</div>\s*([0-9:]+)', html)
        finish = match.group(1).strip() if match else "Sin tiempo"
    except:
        finish = "Sin tiempo"

    resultados.append({
        "place_overall": place_overall,
        "place_gender": place_gender,
        "nombre": nombre,
        "pais": pais,
        "category": categoria,
        "event": event,
        "finish": finish
    })

driver.quit()

df = pd.DataFrame(resultados)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import re

def scrape_london_test(gender, event_name, url):
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(options=options)

    print(f"\n🔍 Scrapeando 10 atletas – {event_name} – Género: {gender}")
    driver.get(url)
    time.sleep(4)

    resultados = []
    atletas = driver.find_elements(By.CSS_SELECTOR, "li.list-group-item.row")[:10]

    for atleta in atletas:
        try:
            place_overall = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-secondary").text
        except:
            place_overall = "Sin posición general"

        try:
            place_gender = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-primary").text
        except:
            place_gender = "Sin posición género"

        try:
            name_pais = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-fullname a").text
            nombre = re.sub(r"\s*\([A-Z]{3}\)", "", name_pais).strip()
            pais_match = re.search(r"\(([A-Z]{3})\)", name_pais)
            pais = pais_match.group(1) if pais_match else "Sin país"
        except:
            nombre = "Sin nombre"
            pais = "Sin país"

        try:
            categoria = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-age_class").text.replace("Category", "").strip()
        except:
            categoria = "Sin categoría"

        try:
            finish_block = atleta.find_element(By.XPATH, ".//div[contains(@class, 'type-time') and .//div[text()='Finish']]")
            html = finish_block.get_attribute("innerHTML")
            match = re.search(r'</div>\s*([0-9:]+)', html)
            finish = match.group(1).strip() if match else "Sin tiempo"
        except:
            finish = "Sin tiempo"

        resultados.append({
            "place_overall": place_overall,
            "place_gender": place_gender,
            "nombre": nombre,
            "pais": pais,
            "category": categoria,
            "event": event_name,
            "finish": finish,
            "gender": gender
        })

    driver.quit()
    return pd.DataFrame(resultados)

# URLs por categoría y género
urls = {
    "Elite_M": ("Elite", "M", "https://results.tcslondonmarathon.com/2024/?event=ELIT&num_results=1000&pid=search&search%5Bsex%5D=M&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex"),
    "Elite_W": ("Elite", "W", "https://results.tcslondonmarathon.com/2024/?event=ELIT&num_results=1000&pid=search&search%5Bsex%5D=W&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex"),
    "Mass_M": ("Mass", "M", "https://results.tcslondonmarathon.com/2024/?event=MAS&num_results=1000&pid=search&search%5Bsex%5D=M&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex"),
    "Mass_W": ("Mass", "W", "https://results.tcslondonmarathon.com/2024/?event=MAS&num_results=1000&pid=search&search%5Bsex%5D=W&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex")
}

# Ejecutar todos los casos
dfs = []
for key, (event, gender, url) in urls.items():
    df = scrape_london_test(gender, event, url)
    dfs.append(df)

# Concatenar todo
df_londres = pd.concat(dfs, ignore_index=True)


print("\n✅ Prueba completada para Londres. Archivo: 'prueba_london_2024_10_por_categoria_genero.csv'")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import re
from tqdm import tqdm

def scrape_london_all(gender, event_name, url_base, total_pages, archivo_salida):
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(options=options)
    resultados = []

    for pagina in tqdm(range(1, total_pages + 1), desc=f"Londres {event_name} {gender}", ncols=100):
        url = f"{url_base}&page={pagina}" if "page=" not in url_base else url_base.replace("page=1", f"page={pagina}")
        driver.get(url)
        time.sleep(4)

        atletas = driver.find_elements(By.CSS_SELECTOR, "li.list-group-item.row")

        for atleta in atletas:
            try:
                place_overall = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-secondary").text
            except:
                place_overall = "Sin posición general"

            try:
                place_gender = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-place.place-primary").text
            except:
                place_gender = "Sin posición género"

            try:
                name_pais = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-fullname a").text
                nombre = re.sub(r"\s*\([A-Z]{3}\)", "", name_pais).strip()
                pais_match = re.search(r"\(([A-Z]{3})\)", name_pais)
                pais = pais_match.group(1) if pais_match else "Sin país"
            except:
                nombre = "Sin nombre"
                pais = "Sin país"

            try:
                categoria = atleta.find_element(By.CSS_SELECTOR, ".list-field.type-age_class").text.replace("Category", "").strip()
            except:
                categoria = "Sin categoría"

            try:
                finish_block = atleta.find_element(By.XPATH, ".//div[contains(@class, 'type-time') and .//div[text()='Finish']]")
                html = finish_block.get_attribute("innerHTML")
                match = re.search(r'</div>\s*([0-9:]+)', html)
                finish = match.group(1).strip() if match else "Sin tiempo"
            except:
                finish = "Sin tiempo"

            resultados.append({
                "page": pagina,
                "place_overall": place_overall,
                "place_gender": place_gender,
                "nombre": nombre,
                "pais": pais,
                "category": categoria,
                "event": event_name,
                "finish": finish,
                "gender": gender
            })

        time.sleep(2)

    driver.quit()
    df = pd.DataFrame(resultados)
    df.to_csv(archivo_salida, index=False)
    print(f"\n✅ Guardado parcial en: {archivo_salida}")
    return df

# --- Configuración por grupo ---
jobs = {
    "Elite_M": ("Elite", "M", "https://results.tcslondonmarathon.com/2024/?event=ELIT&num_results=1000&pid=search&search%5Bsex%5D=M&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex", 1, "london_elite_m.csv"),
    "Elite_W": ("Elite", "W", "https://results.tcslondonmarathon.com/2024/?event=ELIT&num_results=1000&pid=search&search%5Bsex%5D=W&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex", 1, "london_elite_w.csv"),
    "Mass_M": ("Mass", "M", "https://results.tcslondonmarathon.com/2024/?page=1&event=MAS&num_results=1000&pid=search&search%5Bsex%5D=M&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex", 31, "london_mass_m.csv"),
    "Mass_W": ("Mass", "W", "https://results.tcslondonmarathon.com/2024/?page=1&event=MAS&num_results=1000&pid=search&search%5Bsex%5D=W&search%5Bage_class%5D=%25&search%5Bnation%5D=%25&search_sort=place_nosex", 23, "london_mass_w.csv")
}

# Ejecutar y combinar
dfs = []
for key, (event, gender, url, total_pages, output_file) in jobs.items():
    df = scrape_london_all(gender, event, url, total_pages, output_file)
    dfs.append(df)

# Concatenar y exportar
df_londres = pd.concat(dfs, ignore_index=True)
df_londres.to_csv("../data/resultados_london_2024_completo.csv", index=False)
print("\n✅ LONDRES 2024 completado. Guardado en: 'london_2024_completo.csv'")
```
