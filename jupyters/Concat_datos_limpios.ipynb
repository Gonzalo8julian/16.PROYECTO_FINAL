```python
df_berlin = pd.read_excel('../data/datos_limpios/resultados_berlin_2024.xlsx')
df_berlin.head()
```

```python
df_boston = pd.read_excel('../data/datos_limpios/resultados_boston_2024.xlsx')
df_boston.head()
```

```python
df_chicago = pd.read_excel('../data/datos_limpios/resultados_chicago_2024.xlsx')
df_chicago.head()
```

```python
df_london = pd.read_excel('../data/datos_limpios/resultados_london_2024.xlsx')
df_london
```

```python
#concatenamos los dataframes en uno solo llamado df_six_majors:
df_six_majors = pd.concat([df_berlin, df_boston, df_chicago, df_london], ignore_index=True)
df_six_majors
```

```python
df_six_majors.info()
```

```python
#Lo guardamos en un archivo excel llamado resultados_six_majors_2024.xlsx
df_six_majors.to_excel('../data/datos_limpios/resultados_six_majors_2024.xlsx', index=False)
#Lo guardamos en un archivo csv llamado resultados_six_majors_2024.csv
df_six_majors.to_csv('../data/datos_limpios/resultados_six_majors_2024.csv', index=False)
```
